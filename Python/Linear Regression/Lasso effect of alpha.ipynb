```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso
from sklearn.datasets import make_regression
```

```
df = make_regression(n_features=5, n_informative=3)
```

```
# Data for dataframe
df2 = pd.DataFrame(df[0])
df2['y'] = df[1]

# X and Y
X = df2.drop('y', axis=1)
y = df2.y

# set of alphas to test
alphas = [0.1, 1, 10, 40, 60, 100]

# loop models
for a in alphas:
  lasso_reg = Lasso(alpha = a)
  lasso_reg.fit(X,y)
  # Predicts
  df2['a'+ str(a)] = lasso_reg.predict(X)
```

```
# Plot with Mosaic
fig = plt.figure(layout= 'constrained', figsize=(18,8))
mosaic = fig.subplot_mosaic('''
                            abc
                            def
                            ''')

# Plot A
mosaic['a'].scatter( data=df2,  x= 'y',   y= 'a0.1')

# Plot B
mosaic['b'].scatter( data=df2,  x= 'y',   y= 'a1' );

# Plot C
mosaic['c'].scatter( data=df2,  x= 'y',   y= 'a10' );

# Plot D
mosaic['d'].scatter( data=df2,  x= 'y',   y= 'a40' );

# Plot E
mosaic['e'].scatter( data=df2,  x= 'y',   y= 'a60' );

# Plot F
mosaic['f'].scatter( data=df2,  x= 'y',   y= 'a100' );


# Define Titles
titles = ['Lasso with alpha=0.1', 'Lasso with alpha=1', 'Lasso with alpha=10', 'Lasso with alpha=40', 'Lasso with alpha=60', 'Lasso with alpha=100']

# Um titulo por plot
for ax, g_title in zip(mosaic.items(), titles):
    ax[1].set_title(g_title, fontstyle='italic')
```
