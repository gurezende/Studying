```
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from mlxtend.plotting import plot_decision_regions
```

```
# Create dataset
data = make_classification(n_features=2,
                         n_informative=2,
                         n_redundant=0,
                         n_repeated=0,
                         class_sep=0.85,)
```

```
# Dataset for visualization
df = pd.DataFrame(data[0], columns= ['v1', 'v2'])
df['y'] = data[1]

# Split X & y
X = data[0]
y = data[1]
```

```
# Dict to store results
results = []

for d in [1,3,5]:
  for c in [0,4,8]:
    for C in [1,100]:
      # Instance model
      clf = SVC(kernel= 'poly', degree=d, coef0=c, C=C)
      # Fit the model
      clf.fit(X,y)

      # Predict
      #df[f'param[d{d}_c{c}_C{C}]'] = clf.predict(X,y)
      results.append([f'[d{d} c{c} C{C}]', clf.score(X,y)])

# View results
pd.DataFrame(tuple(results), columns=['params', 'score'])
```

```
# Instance model
clf = SVC(kernel= 'poly', degree=1, coef0=0, C=1)

# Fit the model
clf.fit(X,y)

# Predict
plt.title('MODEL | SVC(kernel= "poly", degree=1, coef0=0, C=1)')
plot_decision_regions(X, y, clf= clf);
```
